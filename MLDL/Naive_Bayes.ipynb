## Naive Bayes

나이브 베이즈 분류기는 인스턴스의 레이블을 예측할 때 베이즈 정리를 사용하여 판단한다. 

- 베이즈 정리  
: 사전 확률과 사후 확률 간의 관계에 대해서 설명하는 정리로 사전 확률과 우도 확률을 안다면 사후 확률을 알 수 있다는 것이다.  

    1. 사전 확률 : 관측자가 이미 알고 있는 사건으로부터 나온 확률  
    2. 우도 확률 : 이미 알고 있는 사건이 발생했다는 조건 하에 다른 사건이 발생할 확률  
    3. 사후 확률 : 사전확률과 우도확률을 통해 알게되는 조건부 확률
    
![image](https://user-images.githubusercontent.com/69336270/128629925-2224a389-dba5-4722-8db6-f6c72729d840.png)


그렇다면 나이브 베이즈 분류기 앞에는 `나이브`라는 수식어가 붙은 것일까?  

데이터셋 내에 있는 모든 특성이 관계를 맺고 있다면 클래스가 k개이고 특성이 d개일 때 Class conditional density를 구하기 위해서는 $(2^{d}-1)k$ 개의 파라미터가 필요하다. 특성의 개수에 따라 필요한 파라미터의 수가 기하급수적으로 증가하는 만큼 필요한 연산량과 메모리 양 역시 기하급수적으로 증가한다는 문제점이 발생한다.   


이러한 문제를 해결하기 위해서는 나이브 베이즈 분류기에서 사용하는 베이즈 정리에서 모든 사건이 **조건부 독립**이라는 가정을 정의해주어야 한다.

![image](https://user-images.githubusercontent.com/69336270/128630041-d2e28513-3370-405d-becd-774ca69d258f.png)

다음 그림과 같이 사건 A가 모두 독립일 경우 사건 B에 대한 확률은 조건부 독립을 통해 구할 수 있기 때문이다. 이처럼 조건부 독립을 가성 시 특성의 개수를 줄이지 않고도 필요한 파라미터의 개수를 줄일 수 있다. 

하지만 현실세계에서는 특성들이 서로 조건부독립을 만족하지 않는 경우가 대부분이다. 파라미터의 개수를 줄이기 위한 가정이 말 그대로 너무 순진한(Naive, 나이브) 한 가정이기 때문에 이 가정을 사용한 베이즈 분류기를 나이브 베이즈 분류기 라고 부른다. 

- 장점 
    1. 연산량이 작아 간단하고, 빠르며 정확한 모델
    2. 연속형보다 이산형 데이터에서 성능이 좋음
    3. 입력 공간의 차원이 높을 때 유리
    4. 텍스트에서 강점 (단어 혹은 문서 분류)
    
- 단점
    1. 한번도 등장하지 않은 데이터에 대해 취약한 - 해결책 : 라플라스 스무딩
        - 라플라스 스무딩  
            :  likihood가 0이 되는 것을 방지하기 위해 최소한의 확률을 정해주는 기법으로, 실제로 관찰한 것보다 1번 더 봤다고 가정하여 값을 보정해주어 확률이 0일 경우를 제외해준다.
    2. 조건부 독립이라는 가정 자체가 비현실적
