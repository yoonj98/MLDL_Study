## Clustering

- Clustering이란  
: 데이터 중 유사한 속성을 갖는 관측치들을 묶어 몇개의 군집으로 나누는 것
<br><br>  
- 기준
   1. 동일한 군집에 소속된 관측치들은 서로 유사할 수록 좋다.
   2. 서로 다른 군집에 소속된 관측치들은 서로 다를 수록 좋다.
<br><br>  
- 분류 VS 군집화
    - 분류 : 사전 정의된 범주가 있는 데이터로부터 예측 모델을 학습하는 문제 (Supervised + Labeled)
    - 군집화 : 사전 정의된 범주가 없는 데이터에서 최적의 그룹을 찾는 문제 (Unsupervised + Unlabeled)
<br><br>
- 유사도 척도
    1. 유클리디안 거리
    2. 맨해튼 거리
    3. 마할라노비스 거리
    4. 상관계수 거리
<br><br> 
- 군집화 알고리즘 종류
1. 계층적 군집화  
    - 가까운 집단부터 개체를 묶어나가는 방식  
    - 덴드로그램 생성  
2. 분리형 군집화  
    - 전체 데이터 영역을 특정 기준에 의해 동시에 구분  
    - 각 개체들은 사전에 정의된 개수의 군집 중 하나에 속함    
3. 자기조직화 지도  
    - 2차원 격자에 각 개체들이 대응하도록 인공신경망과 유사한 학습을 통해 군집 도츨  
4. 밀도 기반 군집화  
    - 데이터 분포를 기반으로 고밀도 영역으로 전체 영역을 구분
<br><br>
- 계층적 군집화
1. 계층적 트리모형을 통해 개별 개체들을 계층적으로 유사한 군집과 통합한다.
2. 덴드로그램을 통해 시각화가 가능
3. 사전 군집 수를 정하지 않아도 수행 가능 - 덴드로그램 생성 후 적절 수준에서 커트 <br><br>
![image](https://user-images.githubusercontent.com/69336270/128631226-6c4d3e26-09d9-4cf0-8db3-0a8d990f10c4.png)

<과정>  
1. 모든 개체 간의 거리에 대한 유사도 행렬 계산  
2. 거리가 인접한 관측치끼리 군집 형성  
3. 유사도 행렬 업데이트  
4. 위 과정 반복  

- K-means 군집화
1. 대표적 분리형 군집화 알고리즘
2. 각 군집은 하나의 중심을 갖는다.
3. 각 개체는 가장 가까운 중심에 할당되며, 같은 중심에 할당된 개체들이 모여 하나의 군집을 형성한다. 
4. 사전에 군집 개수 K가 정해져야한다.
  
<과정>    
1. 초기 중심 K개를 무작위로   생성 
2. 개별 관측치로부터 각 중심까지의 거리 계산 후, 가장 가까운 중심이 이루는 군집에 관측치 할당    
3. 각 군집의 중심을 다시 계산  
4. 중심이 변하지 않을 때까지 반복  
      
<단점>  
    1. 서로 다른 크기의 군집을 찾지 못함  
    2. 서로 다른 밀도의 군집을 찾지 못함  
    3. 지역적 패턴이 존재하는 군집을 판별하기 어려워 함

- 밀도 기반 군집화  
: 공간 상에서 높은 밀도를 가지고 모인 관측치를 하나의 그룹으로 간주하고 낮은 밀도를 가지고 홀로 있는 관측치를 이상치 혹은 잡음으로 분류하는 알고리즘    
1. 클러스터 개수를 미리 지정할 필요가 없음  
2. 복잡한 형상도 찾을 수 있으며, 어떤 클래스에도 속하지 않는 포인트를 구분할 수 있음  
3. 비교적 느리지만 큰 데이터 셋에서도 사용 가능
